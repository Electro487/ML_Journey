```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.8, 4.1, 6.0, 7.9, 10.2]) 
```

```python
plt.scatter(x, y)
plt.xlabel("Input (x)")
plt.ylabel("Output (y)")
plt.show()
```

```python
# m = 2
# c = 0
m = 1.0
c = 0.2
y_pred = m * x + c
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
errors = y - y_pred
mse = np.mean(errors ** 2)
print("Mean Squared Error:", mse)
```

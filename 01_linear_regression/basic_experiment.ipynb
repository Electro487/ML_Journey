```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
plt.scatter(x, y)
plt.xlabel("Input (x)")
plt.ylabel("Output (y)")
plt.show()
```

```python
m = 2
c = 0
y_pred = m * x + c
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
